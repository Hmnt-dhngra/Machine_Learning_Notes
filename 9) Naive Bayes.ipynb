Naive Bayes is a **probabilistic model** which is used to solve **classification** (both binary and multi class) problem. 

**Output needed** : We need probabilistic reasoning. 

This is based on Baye's theorem => probabilities of an event given the other has happened. **Dependant probabilities**

**Variants of Naive Bayes**

1) **Bernoulli NB** : Whenever your independant features are following a bernoulli distribution, then we need to use Bernoulli NB for the classification. The Bernoulli distribution is a discrete probability distribution that models a single binary experimentâ€”one trial with exactly two possible outcomes.

2) **Multinomial NB** : Whenever your input data is in the form of text, you can use this for ex: Spam classificiation, etc.  

3) **Gaussian NB** : If the features are following gaussian distribution, then we use this Gaussian Naive Bayes for classfication. A Gaussian distribution, also called the Normal distribution, is a continuous probability distribution that describes data which tends to cluster symmetrically around a central value (the mean), with frequencies tapering off equally on both sides. **Bell curve** Dataset needs to be **continuous**.
