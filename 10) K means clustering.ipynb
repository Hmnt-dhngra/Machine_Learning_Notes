## K- means Clustering

K-means clustering is an **unsupervised machine learning algorithm** used to partition data into K distinct, non-overlapping groups (clusters) based on similarity. 


K-means is used when You have unlabeled data with **No target variable** and the **Goal** is to discover structure or patterns

**Core Idea** : 

Given a dataset with no labels, K-means aims to:
- Divide the data into K clusters
- Ensure that points within a cluster are as similar as possible
- Ensure that clusters are as different as possible

Similarity is typically measured using **Euclidean distance.**

Each cluster is represented by a centroid (mean of the points in that cluster).

The algorithm iteratively assigns points to the nearest centroid and then updates the centroids.

**How K-means works (step-by-step)**

1) Choose K (number of clusters)
2) Initialize K centroids (randomly or using k-means++)
3) Assign each data point to the nearest centroid
4) Recompute centroids as the mean of assigned points

- Centroids stop changing significantly, or
- A maximum number of iterations is reached
