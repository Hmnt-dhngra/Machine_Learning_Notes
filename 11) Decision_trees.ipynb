#### **Decision Trees**

**Decision tree** is a **supervised learning algorithm** used for both **classification and regression** tasks. 

It has a hierarchical tree structure which consists of a root node, branches, internal nodes and leaf nodes. It works like a flowchart help to make decisions step by step where:

- Internal nodes represent attribute tests
- Branches represent attribute values
- Leaf nodes represent final decisions or predictions.

```python
from IPython.display import Image, display
display(Image(filename="photos/decision Tree.png"))
```

- The node where we start the question is called root node
- Further, we have decision nodes & ultimately leaf notes (decision nodes)
- Decision node is where we take the decision
- Leaf node is where we get the answer

##### **How Does a Decision Tree Work**
It works by repeatedly splitting the data into smaller and purer groups based on feature values, forming a structure that looks like an upside-down tree.

```python
from IPython.display import Image, display
display(Image(filename="photos/decision Tree2.png"))
```

**1. Root Node (Income)**

First Question: "Is the persons income greater than $50,000?"

- If Yes, proceed to the next question.
- If No, predict "No Purchase" (leaf node).

**2. Internal Node (Age):**

If the persons income is greater than $50,000, ask: "Is the persons age above 30?"

- If Yes, proceed to the next question.
- If No, predict "No Purchase" (leaf node).

**3. Internal Node (Previous Purchases):**

- If the person is above 30 and has made previous purchases, predict "Purchase" (leaf node).
- If the person is above 30 and has not made previous purchases, predict "No Purchase" (leaf node).


##### **General Intuition & notes**
The problems which we try to solve by Decision trees can typically be solved using multiple if & else blocks but that usually will be very very long. 

The decision flow in decision trees is actually conditional control statements. Every node relates to an if-else logic check and answering. 

**Ques :** If decision tree is simply a bunch of if-else statements , why did we really need an ML algorithm for this? 

**Answer:** In real life, we don't know which is gonna be a root node & which node is going to be broken down into branches & how many branches hence we don't know the starting points of the branching. This is taken care by ML algorithms. 
Further, real life datasets have millions of rows and numerous columns, so it will be cumbersome without an ML algorithm to solve this problem. 

##### **Attribute Selection measures**

- Impurity : Impurity is a measure of how mixed or non-homogeneous the target classes are within a node of a decision tree. Generally speaking, it tells u **How good a split is ?**
    - If all samples belong to one class -> The node is pure . A pure node requires **no further splitting**
    - If samples are evenly split across classes -> The node is highly impure

        - 100 % Yes & 0% No --> 0 impurity i.e. Pure
        - 90% Yes & 10% No --> Low impurity
        - 50% Yes & 50% No --> Maximum Impurity

- Entropy : Entropy measures uncertainty, disorder, or impurity in a dataset.
    - Low entropy  data is pure (mostly one class)
    - High entropy  data is mixed (classes evenly distributed)


$$
\text{Entropy}(S) = - \sum_{i=1}^{k} p_i \log_2(p_i)
$$

- \( p_i \) = proportion of class \( i \) in dataset \( S \)

For the set X = {a,a,a,b,b,b,b,b}
- Total instances: 8
- Instances of b: 5
- Instances of a: 3

Entropy H(X)
$$
H(X)
= - \left[
\left(\frac{3}{8}\right)\log_2\left(\frac{3}{8}\right)
+
\left(\frac{5}{8}\right)\log_2\left(\frac{5}{8}\right)
\right]
$$

Substituting the numerical values:

$$
= - \left[
0.375(-1.415) + 0.625(-0.678)
\right]
$$

$$
= - (-0.53 - 0.424)
$$

$$
= 0.954
$$

**1) Information Gain** : It tells us how useful a question (or feature) is for splitting data into groups. 
- It measures how much the uncertainty (entropy) is reduced after the split of a feature. 
- A good question will create clearer groups and the **feature with the highest Information Gain is chosen to make the decision**
- Higher Information gain => Better split
    - We want the splits to be as dissimilar as possible 
$$
\text{Information Gain}(S, A) = \text{Entropy of parent}(S)
- \sum_{v \in \text{Values}(A)} \frac{|S_child|}{|S|} \, \text{Entropy}(S_child)
$$


Example:
Gender = [M, F, F, F, M, F, F, M, F] : High entropy

If we split it into G1 & G2

G1 = [M, F, F, F, F]

G2 = [M, M, F]

Impurity has reduced versus the original group. This may not be having the least impurity buit is definitely lesser than the original. 



##### **Building Decision Tree using Information Gain**
Steps

1) Calculate Parent Entropy (Ep)
2) Try each feature & find entropy of each feature (X1, X2, X3, etc.) as E1, E2, E3, etc. 
3) Check info grain for each feature
    - Ep - E1 => IG1
    - Ep - E2 => IG2

    Whichever info gain is higher i.e. the one having less uncertainty becomes the root node.
