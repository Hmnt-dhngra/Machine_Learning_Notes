## ROC Curve (Receiver Operating characteristics)

- A classifier usually gives a score which actually is a probability having range of 0 to 1 where 1 is positive and 0 negative. 
- To make a decision, we need a threshold. If the score output is > threshold then the class selected will be 1 else 0. 


Flow :  For every entity you get a model score (probability), then basis an assumed threshold you categorise a particular entity in a label bucket. 
In order to know what the assumed threshold should be, you find a ROC curve which is nothing but a curve plotting TPR, FPR for multiple values of thresholds. 

**ROC curve is TPR versus FPR for all thresholds**

**For ROC, you need TPR and FPR**
- TPR: True positive Rate , 
- TPR = TP/(TP+FN)   
ex: Spam classification: How many actual spams were caught successfully is given by TPR i.e. True positive Rate

- FPR = False Positive Rate
- FPR = FP/(FP+TN) , 
How many non-spams were wrongly marked as spam. 



    - True positive means model predicts positive and actual value is also positive
    - True Negative means model predicts negative and actual value is negative
    - False Positive means model predicts positive and actual value is negative
    - False Negative means model predicts negative and actual value is positive

on the ROC, we decid ethe threshold by taking the point where TPR>FPR
